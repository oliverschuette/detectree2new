```python
# Danum
import rasterio
import geopandas as gpd

site_path = base_dir + "/data/Danum_lianas"

img_path = site_path + "/rgb/2017_50ha_Ortho_reproject.tif"
crown_path = site_path + "/crowns/Danum_lianas_full2017.gpkg"
```

```python
# BCI
import rasterio
import geopandas as gpd

site_path = base_dir + "/data/BCI_50ha"

img_path = site_path + "/rgb/Orthomosaic_20200801_BCI_50ha_geo.tif"
crown_path = site_path + "/crowns/010820_validated/Crowns_2020_08_01_MergedWithPlotData.shp"
```

```python
# Here, we set the name of the output folder.
# Set tiling parameters
buffer = 30
tile_width = 40
tile_height = 40
threshold = 0.6
appends = str(tile_width) + "_" + str(buffer) + "_" + str(threshold)

out_dir = site_path + "/tilesClass_" + appends + "/"

# Read in the tiff file
data = rasterio.open(img_path)

# Read in crowns (then filter by an attribute?)
crowns = gpd.read_file(crown_path)
crowns = crowns.to_crs(data.crs.data)
print(crowns.head())
```

```python
def add_crown_classes(crowns):
  """ functions to add a column to the crowns Geodataframe which record the liana conditions when the conditions range from 0 to 100"""
    # add a column which shows the status of a tree
  crowns = crowns.assign(status='clean')

  for i in range (len(crowns)):
    if crowns["LI_2017"][i] > 0:
      crowns["status"][i] = "infested"
    elif crowns["LI_2017"][i] != crowns["LI_2017"][i]:
      crowns['status'][i] = "delete"
    crowns = crowns.drop(crowns[crowns.status == "delete"].index)
  return(crowns)

crowns = add_crown_classes(crowns)

#now see if the function behaves as you want
print(crowns.head(10))
```

```python
from detectree2.preprocessing.tiling import record_classes, tile_data, to_traintest_folders, load_class_mapping
import pickle
import os
import json
import time
import shutil

# Remove existing tile directory
shutil.rmtree(out_dir, True)

class_column = 'status'
#class_column = None

# Record the classes and save the class mapping
record_classes(
    crowns=crowns,  # Geopandas dataframe with crowns
    out_dir=out_dir,  # Output directory to save class mapping
    column=class_column,  # Column used for classes
    save_format='json'  # Choose between 'json' or 'pickle'
)

# Load the class-to-index mapping from the recorded classes
#class_mapping_file = os.path.join(out_dir, "class_to_idx.json")

#with open(class_mapping_file, 'r') as f:
#    class_to_idx = json.load(f)


start_time = time.time()

# Perform the tiling, ensuring the selected class column is used
tile_data(
    img_path=img_path,
    out_dir=out_dir,
    buffer=buffer,
    tile_width=tile_width,
    tile_height=tile_height,
    crowns=crowns,
    threshold=threshold,
    class_column=class_column  # Use the selected class column (e.g., 'species', 'status')
    #class_mapping_file=class_mapping_file  # Load the class-to-index mapping
)

end_time = time.time()
print(f"Time taken to tile: {end_time - start_time} seconds")

# Split the data into training and validation sets
to_traintest_folders(
    tiles_folder=out_dir,  # Directory where tiles are saved
    out_folder=out_dir,  # Final directory for train/test data
    test_frac=0,  # Fraction of data to be used for testing
    folds=5,  # Number of folds (optional, can be set to 1 for no fold splitting)
    strict=False,  # Ensure no overlap between train/test tiles
    seed=42  # Set seed for reproducibility
)
```
