```python
import rasterio
### PARACOU 2016
site_path = "/content/drive/Shareddrives/detectree2/data/Paracou"
img_path = site_path + "/rgb/2016/Paracou_RGB_2016_10cm.tif"
tiles_path = site_path + "/tilespred/"
model_path = "/content/drive/Shareddrives/detectree2/models/220629_ParacouSepilokDanum_JB.pth"

# Read in the tiff file
data = rasterio.open(img_path)
```

```python
from detectree2.models.reproject import reproject_to_geojson_spatially
reproject_to_geojson_spatially(data, tiles_path + "predictions_geo1/", tiles_path + "predictions/", EPSG = "32622")
```

```python
from detectree2.models.predict import stitch_crowns
folder = "/content/drive/Shareddrives/detectree2/data/Paracou/tilespred/predictions_geo"
crowns = stitch_crowns(folder, 1)
print(crowns)
crowns.to_file("/content/drive/Shareddrives/detectree2/data/Paracou/crowns_out.gpkg")
```
